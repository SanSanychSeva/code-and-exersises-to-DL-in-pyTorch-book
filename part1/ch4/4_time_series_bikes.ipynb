# считываем данные в нампи, и не меняя хранилища создаем объект тензора

```python
bikes_numpy = np.loadtxt(
    "data/bike-sharing-dataset/hour-fixed.csv", 
    dtype=np.float32, 
    delimiter=",", 
    skiprows=1, 
    converters={1: lambda x: float(x[8:10])})              # сразу вычленяет порядковый номер дня в месяце 
```

```python
bikes = torch.from_numpy(bikes_numpy)
```

```python
bikes.shape, bikes.stride()
```

# далее, зная, что данные уже упорядочены и пропуски заполнены нулевыми арендами, выделяем отдельное измерение под сутки: Дни - Каналы - Часы (N-C-L)

```python
daily_bikes = bikes.view(-1, 24, bikes.shape[1])
daily_bikes.shape, daily_bikes.stride()            # N-L-C
```

```python
daily_bikes = daily_bikes.transpose(1, 2)
daily_bikes.shape, daily_bikes.stride()            # N-C-L
```

**NB!:** при вызове метода `view` на тензоре возвращает новый тензор с другими размерностью и шагами без изменения хранилища. 
Это позволяет перегруппировывать тензор практически без затрат, поскольку никакие данные копировать не нужно. Для вызова `view` нам пришлось указать новую форму возвращаемого тензора. `-1` здесь играет роль «заполнителя», означающего «столько индексов, сколько остается с учетом прочих измерений и исходного количества элементов»

## for one day

```python
first_day = bikes[:24].long()
weather_onehot = torch.zeros(first_day.shape[0], 4)
first_day[:,9]
```

```python
weather_onehot.scatter_(
    dim=1, 
    index=first_day[:,9].unsqueeze(1).long() - 1, # поскольку погодные условия варьируются от 1 до 4, а индексы начинаются с 0
    value=1.0)
```

```python
torch.cat((bikes[:24], weather_onehot), 1)[:1]
```

# делаем то же самое с тензором измененной формы daily_bikes, его форма (B, C, L), где L = 24. 
## Сначала создаем нулевой тензор с теми же B и L, но с числом дополнительных столбцов C

```python
daily_weather_onehot = torch.zeros(daily_bikes.shape[0], 4,
                                   daily_bikes.shape[2])
daily_weather_onehot.shape
```

```python
daily_weather_onehot.scatter_(
    1, daily_bikes[:,9,:].long().unsqueeze(1) - 1, 1.0)
daily_weather_onehot.shape
```

## И производим конкатенацию по измерению C

```python
daily_bikes = torch.cat((daily_bikes, daily_weather_onehot), dim=1)
```

## далее варианты нормализации

```python
daily_bikes[:, 9, :] = (daily_bikes[:, 9, :] - 1.0) / 3.0
```

```python
temp = daily_bikes[:, 10, :]
temp_min = torch.min(temp)
temp_max = torch.max(temp)
daily_bikes[:, 10, :] = ((daily_bikes[:, 10, :] - temp_min)
                         / (temp_max - temp_min))
```

```python
temp = daily_bikes[:, 10, :]
daily_bikes[:, 10, :] = ((daily_bikes[:, 10, :] - torch.mean(temp))
                         / torch.std(temp))
```
