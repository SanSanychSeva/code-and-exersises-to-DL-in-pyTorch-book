## кодирование разметки

### номера классов

```python
target = wineq[:, -1].long()
target
```

```python
print('\nFYI: кол-во неправильных округлений при конверсии типов =', ((target - wineq[:, -1]) * (target - wineq[:, -1])).sum().numpy())
```

**Хранение оценок вкусовых качеств вина в целочисленном векторе влечет их упорядоченность, что в данном случае может быть неуместно, ведь оценка 1 ниже оценки 4. Это также подразумевает наличие расстояния между оценками: то есть расстояние между 1 и 3 равно расстоянию между 2 и 4. Если для нашей величины это подходит — замечательно.**

**Если же, напротив, оценки совершенно дискретны, например, означают сорт винограда, то ohe-кодирование подойдет намного лучше,
поскольку не подразумевает упорядоченности или расстояния**

### *ohe* используя метод `tensor.scatter_`

Syntax: `self_tensor.scatter_(axis, index, src)`
* `axis` (int) – the axis along which to index
* `index` (LongTensor) – the indices of elements to scatter, should have same number of dimensionalities as `self_tensor`
* `src` (Tensor) – the source of element(s) to scatter

**параметры избыточны, но это дает дополнительный чек**

```python
target_onehot = torch.zeros(target.shape[0], 10)
target_onehot
```

```python
target_unsqueezed = target.long().unsqueeze(1)
target_unsqueezed
```

```python
target_onehot.scatter_(1, target_unsqueezed, 1.0)
```

**Пояснение:** аргументы метода scatter_ следующие:
* измерение, по которому указываются следующие два аргумента;
* тензор-столбец с индексами заполняемых элементов;
* тензор с элементами для распределения или одно заполняющее скалярное
значение (в данном случае 1).

Другими словами, предыдущий вызов гласит: «Для каждой строки нужно воспользоваться индексом целевой метки (совпадающим с оценкой в нашем случае) в качестве индекса столбца для задания значения 1,0». В результате получается тензор, кодирующий категориальную информацию.

Размерность второго аргумента метода scatter_, тензора индексов, должна совпадать с заполняемым тензором. Поскольку размерность тензора target_onehot равна 2 (4898 × 10), необходимо добавить еще одно фиктивное измерение в target с помощью функции unsqueeze

**в отличие от `inline=True` пандас, в торче self-методы и self меняют (ниже) и его же возвращают (выше)**

# нормализация данных

## вектор средних

```python
data_mean = torch.mean(data, dim=0)
data_mean
```

## вектор дисперсий

```python
data_var = torch.var(data, dim=0)
data_var
```

## нормируем

```python
data_normalized = (data - data_mean) / torch.sqrt(data_var)
```

## проверяем

```python
torch.mean(data_normalized, dim=0)
```

**вот она, точность `float32`!**

```python
torch.var(data_normalized, dim=0)
```

**кажется, что VAR и STD всегда получаются точно 1, а вот средние не так точны, но это трюки округления, которые видны, если вычесть 1.0:**

```python
torch.var(data_normalized, dim=0) - 1.0
```

**а также порядка вычислений, при которых постоянно идут округления и накапливается ошибка:**

```python
data_normalized = data / torch.sqrt(data_var) - data_mean / torch.sqrt(data_var)  # немного другой порядок действий

torch.var(data_normalized, dim=0) - 1.0
```

# NB!
PyTorch дает возможность непосредственно использовать индексы классов в качестве целей при обучении нейронной сети. Однако для использования оценок в качестве категориального входного сигнала сети необходимо преобразовать его в ohe-кодированный тензор.

## сравним статистики хороших, средних и плохих вин - по среднему значению, например

```python
bad_data = data[target <= 3]
mid_data = data[(target > 3) & (target < 7)]
good_data = data[target >= 7]

bad_mean = torch.mean(bad_data, dim=0)
mid_mean = torch.mean(mid_data, dim=0)
good_mean = torch.mean(good_data, dim=0)

for i, args in enumerate(zip(ddf.columns, bad_mean, mid_mean, good_mean)):
    print('{:2} {:20} {:6.2f} {:6.2f} {:6.2f}'.format(i, *args))
```

**монотонность выявляет явную зависимость**

## сделаем пороговый классификатор по одной переменной - для прикола

```python
total_sulfur_threshold = 141.83
total_sulfur_data = data[:,6]
predicted_indexes = torch.lt(total_sulfur_data, total_sulfur_threshold)

predicted_indexes.shape, predicted_indexes.dtype, predicted_indexes.sum()
```

```python
actual_indexes = target > 5

actual_indexes.shape, actual_indexes.dtype, actual_indexes.sum()
```

```python
n_matches = torch.sum(actual_indexes & predicted_indexes).item()
n_predicted = torch.sum(predicted_indexes).item()
n_actual = torch.sum(actual_indexes).item()

n_matches/len(ddf), n_matches / n_predicted, n_matches / n_actual
```

**это знаменитая тройка `(Accuracy, Precision, Recall)` для нашего порогового классификатора, выявляющего плохие вина!**
